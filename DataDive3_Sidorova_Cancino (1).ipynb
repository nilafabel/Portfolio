```python
# Importing the necessary libraries for data visualization and analysis
import matplotlib.pyplot as plt       # Matplotlib
import seaborn as sns                 # Seaborn
import numpy as np                    # NumPy
import pandas as pd                   # Pandas
import matplotlib.ticker as mticker   # Ticker
import matplotlib.patches as mpatches # Patches
```

```python
# Loads datasets into DataFrames
df_Engineer = pd.read_csv('Data_Engineer.csv')
df_Analyst = pd.read_csv('Data_Analyst.csv')
print(df_Engineer.head()) # prints first row
print(df_Analyst.head())  # prints first row
```

```python
merged_df = pd.concat([df_Engineer, df_Analyst]) # Merges datasets by concatenating them
merged_df.to_csv('merged_dataset.csv', index=False) # Saves the merged dataset to a CSV file

# Prints summary statistics of the merged dataset
print(merged_df.describe())
merged_df
```

```python
merged_df = merged_df.drop_duplicates() # Drops duplicate rows 
missing_values = merged_df.isnull().sum() # Calculates the number of missing values in each column
merged_df.dtypes, missing_values # Displays the data types of each column + the count of missing values
```

```python
merged_df.dropna(inplace=True) # Drops rows with any missing values
missing_values_after = merged_df.isnull().sum() # Calculates the number of missing values in each column after dropping rows with missing values
print(missing_values_after)
```

```python
# Lists of columns that need to be converted to string type and have specific values replaced
object_columns_to_replace = [
    'Competitors', 'Revenue', 'Sector', 'Industry', 'Type of ownership', 'Size', 
    'Headquarters', 'Location', 'Company Name', 'Job Description', 
    'Salary Estimate', 'Job Title'
]
merged_df[object_columns_to_replace] = merged_df[object_columns_to_replace].astype(str) # Converts the specified columns to string type

# Replaces the value '-1' with 'Unknown / Non-Applicable' in the specified columns
merged_df[object_columns_to_replace] = merged_df[object_columns_to_replace].replace('-1', 'Unknown / Non-Applicable')
merged_df
```

```python
Num_columns_to_replace = ['Rating', 'Founded'] # Lists of numerical columns where the value -1 needs to be replaced with NaN

# Replaces the value -1 with NaN in the specified numerical columns
merged_df[Num_columns_to_replace] = merged_df[Num_columns_to_replace].replace(-1, np.nan) 
merged_df
```

```python
TF_columns_to_replace = ['Easy Apply'] # Lists of True/False columns where the value -1 needs to be replaced with 'FALSE'
merged_df[TF_columns_to_replace] = merged_df[TF_columns_to_replace].astype(str) # Converts the specified columns to string type

# Replaces the value '-1' with 'FALSE' in the specified columns
merged_df[TF_columns_to_replace] = merged_df[TF_columns_to_replace].replace('-1', 'FALSE')
merged_df
```

```python
keywords = r"(Data Engineer |Data Analyst )" # Defines a pattern that includes the required keywords

# Filters the rows where 'Job Title' contains any of the keywords
merged_df = merged_df[merged_df['Job Title'].str.contains(keywords, case=False, na=False)]
merged_df
```

```python
def extract_salary(salary_estimate):
    cleaned_salary = salary_estimate.replace('$', '').replace('K', '').strip() # Removes unwanted characters and parenthetical text
   
    # Removes any text within parentheses
    if '(' in cleaned_salary:
        cleaned_salary = cleaned_salary.split('(')[0].strip()
        
    salary_parts = cleaned_salary.split('-') # Splits the salary range
   
    # Checks if both parts of the salary range are present and contain valid numbers
    if len(salary_parts) == 2 and salary_parts[0].strip() and salary_parts[1].strip():
        min_salary = int(salary_parts[0].strip()) * 1000
        max_salary = int(salary_parts[1].strip()) * 1000
        return min_salary, max_salary        
    elif len(salary_parts) == 1 and salary_parts[0].strip():
        salary = int(salary_parts[0].strip()) * 1000
        return salary, salary
        
    # Returns None if salary data is missing or improperly formatted
    return None, None
```

```python
# Initializes empty lists to store minimum and maximum salaries
min_salaries = []
max_salaries = []

# Iterates over each salary estimate in the 'Salary Estimate' column
for salary in merged_df['Salary Estimate']:
    
    # Extracts minimum and maximum salary from the salary estimate using a custom function `extract_salary`
    min_salary, max_salary = extract_salary(str(salary))
    
    # Appends the extracted minimum and maximum salaries to their respective lists
    min_salaries.append(min_salary)
    max_salaries.append(max_salary)

# Adds the extracted minimum salaries as a new column 'Min Salary' 
merged_df['Min Salary'] = min_salaries

# Adds the extracted maximum salaries as a new column 'Max Salary' 
merged_df['Max Salary'] = max_salaries

# Calculates the average salary and add it as a new column 'Avg Salary'
merged_df['Avg Salary'] = (merged_df['Min Salary'] + merged_df['Max Salary']) / 2
merged_df
```

```python
# Defines a function to categorize job titles
def categorize_job_title(job_title):
    if 'Data Analyst' in job_title:
        return 'Data Analyst'
    elif 'Data Engineer' in job_title:
        return 'Data Engineer'
    else:
        return None  # Excludes any other job titles

# Applys the categorization function to the 'Job Title' column and updates the column
merged_df['Job Title'] = merged_df['Job Title'].apply(categorize_job_title)

merged_df = merged_df.dropna(subset=['Job Title']) # Drops rows where 'Job Title' is NaN

# Groups by 'Job Title' and calculate the average salary for each group
avg_salary_by_group = merged_df.groupby('Job Title')['Avg Salary'].mean().reset_index()

#Bar plot
custom_palette = ['dimgrey', 'hotpink']
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(data=avg_salary_by_group, x='Job Title', y='Avg Salary', palette=custom_palette)

# Creates a custom legend with colors corresponding to each group
handles = [
    mpatches.Patch(color=custom_palette[0], label='Data Analyst'),
    mpatches.Patch(color=custom_palette[1], label='Data Engineer')
]

plt.legend(handles=handles)
plt.title('Average Salary by Job Title Group', fontsize=16)
plt.xlabel('Job Title', fontsize=12)
plt.ylabel('Average Salary ($)', fontsize=12)
plt.show()
avg_salary_by_group
```

```python
# Groups the data to calculate the average rating for each combination of Job Title Group and Industry
avg_ratings = merged_df.groupby(['Industry', 'Job Title'], as_index=False)['Rating'].mean()

# Bar plot
plt.figure(figsize=(12, 12))
custom_palette = ['dimgrey', 'hotpink'] 
sns.barplot(data=avg_ratings, x='Rating', y='Industry', hue='Job Title', palette=custom_palette)
plt.title('Average Rating by Job Title Group', fontsize=16)
plt.xlabel('Average Rating', fontsize=12)
plt.ylabel('Industry', fontsize=12)
plt.legend()
plt.show()
```

```python
data_analyst_ratings = avg_ratings[avg_ratings['Job Title'] == 'Data Analyst'] # Filters `avg_ratings` to get only 'Data Analyst' job titles
data_engineer_ratings = avg_ratings[avg_ratings['Job Title'] == 'Data Engineer'] # Filters `avg_ratings` to get only 'Data Engineer' job titles

# Gets the top 5 'Data Analyst' ratings by sorting and selecting the largest values in the 'Rating' column
top_data_analyst_ratings = data_analyst_ratings.nlargest(5, 'Rating')

# Gets the top 5 'Data Engineer' ratings by sorting and selecting the largest values in the 'Rating' column
top_data_engineer_ratings = data_engineer_ratings.nlargest(5, 'Rating')

print("Top 4 Data Analyst Ratings:")
print(top_data_analyst_ratings)
print("\nTop 5 Data Engineer Ratings:")
print(top_data_engineer_ratings)
```

```python
# Gets the lowest 5 'Data Analyst' ratings by sorting and selecting the smallest values in the 'Rating' column
lowest_data_analyst_ratings = data_analyst_ratings.nsmallest(5, 'Rating')

# Gets the lowest 5 'Data Engineer' ratings by sorting and selecting the smallest values in the 'Rating' column
lowest_data_engineer_ratings = data_engineer_ratings.nsmallest(5, 'Rating')

print("Lowest 4 Data Analyst Ratings:")
print(lowest_data_analyst_ratings)
print("\nLowest 4 Data Engineer Ratings:")
print(lowest_data_engineer_ratings)
```

```python
# Extracts the state from the 'Location' column by splitting the string on ',' and taking the last part,
# then stripping any leading or trailing whitespace
merged_df['State'] = merged_df['Location'].str.split(',').str[-1].str.strip()
merged_df
```

```python
# If 'Job Title' contains 'Engineer', assign 'Engineer'; if it contains 'Analyst', assign 'Analyst'; else, assign 'Other'
merged_df['Job Group'] = np.where(
    merged_df['Job Title'].str.contains('Engineer', case=False), 'Engineer',  # Checks if 'Engineer' is in 'Job Title'
    np.where(merged_df['Job Title'].str.contains('Analyst', case=False), 'Analyst', 'Other')  # Checks if 'Analyst' is in 'Job Title'
)
merged_df
```

```python
# Groups the data by 'State' and 'Job Group', then calculate the average salary for each group
salary_comparison = merged_df.groupby(['State', 'Job Group'])['Avg Salary'].mean().reset_index()

#Bar plot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
custom_palette = ['dimgrey', 'hotpink'] 
sns.barplot(data=salary_comparison, x="State", y="Avg Salary", hue="Job Group", palette=custom_palette)
plt.title('Comparison of Average Salary by Location (Data Analyst vs Data Engineer)', fontsize = 16)
plt.xlabel('State', fontsize = 12)
plt.ylabel('Average Salary($)', fontsize = 12)
plt.xticks(rotation=45)
plt.legend(title='Job Group')
plt.tight_layout()
plt.show()
```

```python
size_mapping = { '1 to 50 employees': 'Small', '51 to 200 employees': 'Small', '201 to 500 employees': 'Small', '501 to 1000 employees': 
                'Medium', '1001 to 5000 employees': 'Medium', '10000+ employees': 'Large', 'Unknown': 'Unknown' }
# Maps the size to categories
merged_df['Company Size Category'] = merged_df['Size'].map(size_mapping)
merged_df = merged_df.reset_index(drop=True)

# Sets the figure size
plt.figure(figsize=(12, 6))

# Defines the order for the x-axis categories
size_order = ['Small', 'Medium', 'Large']

# Strip plot
custom_palette = ['dimgrey', 'hotpink', 'blue']
sns.stripplot(x='Company Size Category', y='Min Salary', data=merged_df, order=size_order, hue='Job Group', palette=custom_palette, dodge=True)
plt.title('Distribution of Minimum Salary by Company Size and Job Title', fontsize=16)
plt.xlabel('Company Size', fontsize=12)
plt.ylabel('Minimum Salary($)', fontsize=12)
plt.legend(title='', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust the legend position
plt.tight_layout()
plt.show()
```

```python
# Gets the counts of 'Easy Apply' values for 'Data Analyst' job titles
analyst_counts = merged_df[merged_df['Job Title'] == 'Data Analyst']['Easy Apply'].value_counts()

# Gets the counts of 'Easy Apply' values for 'Data Engineer' job titles
engineer_counts = merged_df[merged_df['Job Title'] == 'Data Engineer']['Easy Apply'].value_counts()

# Pie Charts 
fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].pie(analyst_counts, labels=analyst_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
axes[0].set_title('Easy Apply for Data Analyst')  # Set the title of the first subplot
axes[1].pie(engineer_counts, labels=engineer_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
axes[1].set_title('Easy Apply for Data Engineer')  # Set the title of the second subplot
plt.show()
```
